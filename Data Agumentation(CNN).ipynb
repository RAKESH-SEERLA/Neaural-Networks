```python
x,y=[],[]
for flower_name,images in flowers_image_dict.items():#"roses":list(data_dir.glob("roses/*"))
    for image in images:
        img=cv2.imread(str(image))
        resized_img=cv2.resize(img,(180,180))
        x.append(resized_img)
        y.append(flowers_labels_dict[flower_name])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
print("x_train",x_train.shape)
print("y_train",y_train.shape)
print("x_test",x_test.shape)
print("y_test",x_test.shape)
```

```python
#Scalling the data
```

```python
x_train=x_train/255
x_test=x_test/255
x_train[0]#successfully scaled the data
```

```python
#Building the CNN model
```

```python
#parameters
#filters
#kernal_size,
#pooling,
#activation
#input_shape
#padding
```

```python
data_agumentation=keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal",input_shape=(180,180,3)),
    tf.keras.layers.RandomRotation(0.1),
    tf.keras.layers.RandomZoom(0.1)
])
```

```python
model=keras.Sequential([
    data_agumentation,
    #CONVOLUTION NEAURAL NETWORK
    keras.layers.Conv2D(filters=16,kernel_size=(3,3),padding="same",activation="relu"),
    keras.layers.MaxPooling2D((2,2)),
    keras.layers.Conv2D(filters=32,kernel_size=(3,3),padding="same",activation="relu"),
    keras.layers.MaxPooling2D((2,2)),
    keras.layers.Conv2D(filters=64,kernel_size=(3,3),padding="same",activation="relu"),
    keras.layers.MaxPooling2D((2,2)),


    #flatten the layyers
    keras.layers.Flatten(),
    #Neaural Networks
    keras.layers.Dense(128,activation="relu"),
    keras.layers.Dense(50,activation="relu"),
    keras.layers.Dense(5,activation="softmax")
    
    
])
```

```python
model.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
```

```python
print("Num GPUs Available: ", len(tf.config.experimental.list_physical_devices('GPU')))
```

```python
model.fit(x_train,y_train,batch_size=32,epochs=10)
```

```python
model.evaluate(x_train,y_train)
```

```python
model.evaluate(x_test,y_test)
```
